```python
#importando las bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
#importando el dataset
store_data = pd.read_csv('store_data.csv',header=None)
```

```python
#mostrando el top 5 de datos
store_data.head()
```

```python
#Convirtiendo nuestro marco de datos de pandas en una lista de listas
records = []
for i in range(0, 7501):
    records.append([str(store_data.values[i,j]) for j in range(0, 20)])
```

```python
#Aplicación de Apriori
#Procesamiento de datos
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
#mostrando el número de asociasiones
print(len(association_results))
```

```python
#Visualizando los resultados
print(association_results[0])
```

```python
#Mostrando los resultados de manera más entendible
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Regla: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Soporte: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confianza: " + str(item[2][0][2]))
    print("Ascensor: " + str(item[2][0][3]))
    print("________________________________________")
```
